# Integrals and Derivatives

```python
x = np.linspace(1, 10, 100)
y = 1/x**2 * np.sin(x)
```

```python
plt.plot(x,y)
```

```python
dydx = np.gradient(y, x) # calculate derivative. y is the function and x is the spacing
```

```python
plt.plot(x, y)
plt.plot(x, dydx)
```

```python
np.cumsum(np.array([1,2,3,4]))
```

```python
y_int = np.cumsum(y) * (x[1] - x[0]) #calculate integral of the function y
```

```python
plt.plot(x, y)
plt.plot(x, dydx)
plt.plot(x, y_int)
```

# Question 4:


Arrange the equations:
$$ P/ (A  \sigma \epsilon) = T ^ 4 $$
$$ P/ (A \sigma \epsilon T_0) = \left( \frac{1}{1+e^{-kt}} \right)^4 $$

```python
kt = np.linspace(0, 3, 100)
P = (1 / (1 + np.exp(-kt))) ** 4
```

```python
plt.plot(kt, P)
```

```python
E = np.cumsum(P) * (kt[1] - kt[0])
E
```

```python
plt.plot(kt, P)
plt.plot(kt, E)
```

```python
plt.plot(kt, E)
plt.xlabel('$kt$', fontsize=20)
plt.ylabel(r'$\left(\frac{k}{A \sigma \epsilon T_0^4}\right) E(kt)$', fontsize=20)
```

#### Systems of Equations

$$3x + 2y + z = 4 $$
$$5x - 5y + 4z = 3 $$

# Question 4


```python
N = 1000
x = np.linspace(0,1,N+1)
dx = x[1] - x[0]
```

```python
main_diag = -2*np.ones(N-1)
off_diag = np.ones(N-2)
```

```python
np.diag(main_diag)
```

```python
np.diag(off_diag, 1)
```

```python
derivative_matrix = (np.diag(main_diag) + np.diag(off_diag, -1) + np.diag(off_diag, 1)) / dx**2
x2_matrix =  np.diag((10*x[1:-1])**2)
LHS_matrix = derivative_matrix + x2_matrix
```

```python
LHS_matrix
```

```python
w, v = np.linalg.eigh(LHS_matrix)
```

```python
w[-1]
```

```python
plt.plot(v[:,-1])
plt.plot(v[:,-2])
plt.plot(v[:,-3])
plt.plot(v[:,-4])
```
