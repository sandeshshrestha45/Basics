# Integrals and Derivatives

```python
x = np.linspace(1, 10, 100)
y = 1/x**2 * np.sin(x)
```

```python
plt.plot(x,y)
```

```python
dydx = np.gradient(y, x) # calculate derivative. y is the function and x is the spacing
```

```python
plt.plot(x, y)
plt.plot(x, dydx)
```

```python
np.cumsum(np.array([1,2,3,4]))
```

```python
y_int = np.cumsum(y) * (x[1] - x[0]) #calculate integral of the function y
```

```python
plt.plot(x, y)
plt.plot(x, dydx)
plt.plot(x, y_int)
```

# Question 4:


Arrange the equations:
$$ P/ (A  \sigma \epsilon) = T ^ 4 $$
$$ P/ (A \sigma \epsilon T_0) = \left( \frac{1}{1+e^{-kt}} \right)^4 $$

```python
kt = np.linspace(0, 3, 100)
P = (1 / (1 + np.exp(-kt))) ** 4
```

```python
plt.plot(kt, P)
```

```python
E = np.cumsum(P) * (kt[1] - kt[0])
E
```

```python
plt.plot(kt, P)
plt.plot(kt, E)
```

```python
plt.plot(kt, E)
plt.xlabel('$kt$', fontsize=20)
plt.ylabel(r'$\left(\frac{k}{A \sigma \epsilon T_0^4}\right) E(kt)$', fontsize=20)
```
